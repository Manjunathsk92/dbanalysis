```python
from itertools import product
```

```python
gener = product([-1,0,1],repeat = 9)
```

```python
train = [lst for lst in gener]
```

```python
new_train = []
for lst in train:
    sum_0s = 0
    sum_1s = 0
    for i in lst:
        if i==-1:
            sum_0s+=1
        if i==1:
            sum_1s+=1
    if abs(sum_0s - sum_1s) <= 1:
        new_train.append(lst)
```

```python
len(new_train)
```

```python
train = new_train
```

```python
def score(lst):
    combins = [[0,1,2],[3,4,5],[6,7,8],\
            [0,3,6],[1,4,7],[2,5,8],
            [0,4,8],[2,4,6]]
    a_win = 0
    b_win = 0
    for comb in combins:
        a_tot=0
        b_tot=0
        
        for i in comb:
            
            if lst[i]==-1:
                a_tot+=1
            if lst[i]==1:
                b_tot+=1
        if a_tot==3:
            a_win+=1
        if b_tot==3:
            b_win+=1
    if a_win >= 1 and b_win >= 1:
        return 'n'
    elif a_win >= 1:
        return 'a'
    elif b_win >= 1:
        return 'b'
    else:
        return 'd'
    
            
    
    
    
```

```python
arr = []
for lst in train:
    result = score(lst)
    if result != 'n':
        if result == 'a':
            arr.append(list(lst)+[1,1])
        elif result == 'b':
            arr.append(list(lst)+[-1,1])
        elif result =='d':
            arr.append(list(lst)+[0,0])
```
