## Creating a 'cluster matrix'
Conatins the mean dwell and travel times for every stop for every hour of the day for every day of the week.

```python
import os
import pandas as pd
from dbanalysis.stop_tools import stop_getter
s_getter = stop_getter()
from dbanalysis import headers as hds
base_dir = '/home/student/data/stops'
time_format = "%d-%b-%y %H:%M:%S"
directories = os.listdir(base_dir)
d={}
count = 0
for stopA in directories:
    d[stopA]={}
    files = [file for file in os.listdir(base_dir+'/'+stopA) if file != 'oprhans.csv']
    for sB in files:
        count+=1
        if count % 100 == 0:
            print(count)
        stopB = sB.split('.')[0]
        d[stopA][stopB]={}
        distance=s_getter.get_stop_distance(stopA,stopB)
        d[stopA][stopB]['distance']=distance
        if d[stopA][stopB]['distance'] == None:
            continue
        
        df = pd.read_csv(base_dir+'/'+stopA+'/'+sB, names=hds.get_stop_link_headers())
        df['dt']=pd.to_datetime(df['dayofservice'],format=time_format)
        df['hour']=df['actualtime_arr_from']//3600
        df['distance']=distance
        df['day']=df['dt'].dt.dayofweek
        df['traveltime']=df['actualtime_arr_to']-df['actualtime_dep_from']
        df['speed']=df['distance'] / df['traveltime']
        df['dwelltime']=df['actualtime_dep_from']=df['actualtime_arr_from']
        for day in range(0,7):
            
            for hour in range(0,24):
                x=df[(df['day']==day) & (df['hour']==hour)]
                if x.shape[0]>1:
                
                    d[stopA][stopB]['speed'+'_'+str(day)+'_'+str(hour)]=x['speed'].mean()
                    d[stopA][stopB]['dwell'+'_'+str(day)+'_'+str(hour)]=x['dwelltime'].mean()
                del(x)
        del(df)
import pickle
with open('clustermatrix.pickle','wb') as handle:
    pickle.dump(d,handle,protocol=pickle.HIGHEST_PROTOCOL)
                
                
        
        
```

```python
with open('/home/student/dbanalysis/dbanalysis/resources/fullclustermatrix.pickle','wb') as handle:
    pickle.dump(d,handle,protocol=pickle.HIGHEST_PROTOCOL)
```

## Well its done. Took 2 hours to compress the data into a 27 megabyte dictionary. 

Now make it into a data frame

```python
frame = {'stopA':[],'stopB':[],'distance':[]}
for day in range(7):
    for hour in range(24):
        frame['speed_'+str(day)+'_'+str(hour)]=[]
        frame['dwell_'+str(day)+'_'+str(hour)]=[]
```

```python
for stop_a in d:
    
    for stop_b in d[stop_a]:
        
        data = d[stop_a][stop_b]
        
    
        
        if len(data)>1:
            frame['distance'].append(data['distance'])
            frame['stopA'].append(stop_a)
            frame['stopB'].append(stop_b)
            for day in range(7):
                
                for hour in range(24):
                    
                    if 'dwell_'+str(day)+'_'+str(hour) in data:
                        frame['dwell_'+str(day)+'_'+str(hour)].append(float(data['dwell_'+str(day)+'_'+str(hour)]))
                    else:
                        frame['dwell_'+str(day)+'_'+str(hour)].append(0)
                    if 'speed_'+str(day)+'_'+str(hour) in data:
                        frame['speed_'+str(day)+'_'+str(hour)].append(float(data['speed_'+str(day)+'_'+str(hour)]))
                    else:
                        frame['speed_'+str(day)+'_'+str(hour)].append(0)
                
            
                        
                        
```

```python
df=pd.DataFrame(frame,columns=frame.keys())
```

### Trying to cluster

```python
from sklearn.cluster import KMeans
```

```python
cols = [col for col in df.columns if col not in ['stopA','stopB','distance']]
```

```python
kmeans = KMeans(n_clusters = 100).fit(df[cols])
```
