```python
def data_frame_contains_route(df,route):
    routes = df['routeid'].unique()
    for r in routes:
        if r.split('_')[0] == route:
            return True
    return False
    
```

```python
def find_missing_links(route_arr,route):
    missing_links = []
    for i in range(1, len(route_arr)-1):
        truth,df= data_frame_exists(route_arr[i],route_arr[i+1],return_df=True)
        if truth and not data_frame_contains_route(df,route):
            missing_links.append(i)
        elif not truth:
            missing_links.append(['x',i])
    return missing_links
    
        
    
    
    
```
