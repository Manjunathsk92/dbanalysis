```python
t=pd.read_csv('/data/gtfs/transfers.txt')
from dbanalysis import stop_tools
sg = stop_tools.stop_getter()
```

```python
t['from_stop_id'] = t['from_stop_id'].apply(lambda x : x[-4:])
```

```python
t['to_stop_id'] = t['to_stop_id'].apply(lambda x : x[-4:])
```

```python
t['from_stop_id'] = t['from_stop_id'].apply(lambda x : x[1:] if x == 0 else x )
```

```python
t['to_stop_id'] = t['to_stop_id'].apply(lambda x : x[1:] if x[0]==0 else x)
```

```python
d={}
for row in t.itertuples():
    if row[1] not in d:
        d[row[1]] = {}
    if row[2] not in d[row[1]]:
        d[row[1]][row[2]] = row[4]
```
